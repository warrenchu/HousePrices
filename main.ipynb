```python
import numpy as np
import pandas as pd
from typing import *
import seaborn as sns
from matplotlib import pyplot as plt
from scipy import stats
```

```python
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
train_df.head()
print(train_df.shape)

# See what has more than 80% NaN values
percent_missing = train_df.isnull().sum()/len(train_df)

# Drop the columns with over 80% NaN
to_be_dropped=['Alley', 'PoolQC', 'Fence', 'MiscFeature']
train_df = train_df.drop(columns=to_be_dropped)
test_df = test_df.drop(columns=to_be_dropped)
```

```python
# Now fill the rest of the NA values
train_df = train_df.fillna(train_df.mean())
test_df = test_df.fillna(test_df.mean())
# Deal with values that are non-numeric
train_df = pd.get_dummies(train_df)
test_df = pd.get_dummies(test_df)
```

```python
# data analysis
sns.distplot(train_df['SalePrice'])

```

```python
# Lot area v. saleprice
plt.plot(train_df["LotArea"], train_df["SalePrice"], 'bo')
plt.xlabel = "LotArea"
plt.ylabel="SalePrice"
plt.show()
```
